```python
def move_forward(position, direction):
    x, y, z = position
    if direction == "N":
        return (x, y + 1, z)
    elif direction == "S":
        return (x, y - 1, z)
    elif direction == "E":
        return (x + 1, y, z)
    elif direction == "W":
        return (x - 1, y, z)
    elif direction == "Up":
        return (x, y, z + 1)
    elif direction == "Down":
        return (x, y, z - 1)

def move_backward(position, direction):
    return move_forward(position, opposite_direction(direction))

def rotate_left(direction):
    if direction == "N":
        return "W"
    elif direction == "W":
        return "S"
    elif direction == "S":
        return "E"
    elif direction == "E":
        return "N"

def rotate_right(direction):
    if direction == "N":
        return "E"
    elif direction == "E":
        return "S"
    elif direction == "S":
        return "W"
    elif direction == "W":
        return "N"

def rotate_up(direction):
    if direction == "N":
        return "Up"
    elif direction == "S":
        return "Down"
    else:
        return direction

def rotate_down(direction):
    if direction == "Up":
        return "N"
    elif direction == "Down":
        return "S"
    else:
        return direction

def opposite_direction(direction):
    if direction == "N":
        return "S"
    elif direction == "S":
        return "N"
    elif direction == "E":
        return "W"
    elif direction == "W":
        return "E"
    elif direction == "Up":
        return "Down"
    elif direction == "Down":
        return "Up"
```
